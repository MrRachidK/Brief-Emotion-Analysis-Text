### Stemming

```python
def stemmer_snowball(text_list):
    snowball = SnowballStemmer(language='english')
    stemmed_list = []
    for liste in text_list:
        stemmed_sentence = []
        for i in liste:
            i = snowball.stem(i)
            stemmed_sentence.append(i)
        stemmed_list.append(stemmed_sentence)
    return stemmed_list

new_list = stemmer_snowball(token_list)
new_list
```
